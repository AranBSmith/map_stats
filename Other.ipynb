```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import re
import statistics


def present(archive_path):
    
    global duty_cycles
    global stance_angles
    global stance_offsets
    global phase_offsets
    
    duty_cycles = []
    stance_angles = []
    stance_offsets = []
    phase_offsets = []
    
    k,_,_,_,_,_,_,z,d1,d2,d3,d4,d5,d6,sa1,sa2,sa3,sa4,sa5,sa6,so1,so2,so3,so4,so5,so6,po1,po2,po3,po4,po5 = np.loadtxt(archive_path, unpack=True)
    
    duty_cycles.append(d1)
    duty_cycles.append(d2)
    duty_cycles.append(d3)
    duty_cycles.append(d4)
    duty_cycles.append(d5)
    duty_cycles.append(d6)
    
    stance_angles.append(sa1)
    stance_angles.append(sa2)
    stance_angles.append(sa3)
    stance_angles.append(sa4)
    stance_angles.append(sa5)
    stance_angles.append(sa6)
    
    stance_offsets.append(so1)
    stance_offsets.append(so2)
    stance_offsets.append(so3)
    stance_offsets.append(so4)
    stance_offsets.append(so5)
    stance_offsets.append(so6)
    
    phase_offsets.append(po1)
    phase_offsets.append(po2)
    phase_offsets.append(po3)
    phase_offsets.append(po4)
    phase_offsets.append(po5)


    print("avg Duty Cycles")
    for i in range(len(duty_cycles)):
        print(statistics.mean(duty_cycles[i]))
    print("avg Stance Angles")
    for i in range(len(stance_angles)):
        print(statistics.mean(stance_angles[i]))
    print("avg Stance Offsets")
    for i in range(len(stance_offsets)):
        print(statistics.mean(stance_offsets[i]))
    print("avg Phase Offsets")
    for i in range(len(phase_offsets)):
        print(statistics.mean(phase_offsets[i]))
```

```python
from sammon import sammon
import matplotlib.pyplot as plt


archive_path = "2307/rhex_text_2019-07-19_15_22_31_12850/archive_2850.dat"
data = np.loadtxt(archive_path)
x = data[:,8:]
target = data[:,7]

[y,E] = sammon.sammon(x, 2, maxiter = 200)

# Plot
fig = plt.figure(figsize=(14,5))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.scatter(y[target >= 0, 0], y[target >= 0, 1], s=20, c='r', marker='x', label="0-1m")
axes.scatter(y[target >= 1, 0], y[target >= 1, 1], s=20, c='y', marker='x', label="1-2m")
axes.scatter(y[target >= 2, 0], y[target >= 2, 1], s=20, c='g', marker='x', label="2-3m")
axes.scatter(y[target >= 3, 0], y[target >= 3, 1], s=20, c='c', marker='x', label="3-4m")
axes.scatter(y[target >= 4, 0], y[target >= 4, 1], s=20, c='b', marker='x', label=">4m")

plt.title('Projections of rhex gaits')
plt.legend(loc=2)
plt.savefig("sammon.pdf")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

archive_path = "2307/rhex_text_2019-07-19_15_22_31_12850/archive_2850.dat"
data = np.loadtxt(archive_path)
x = data[:,8:]

# tripod reference gait control parameters
tripod = np.array([0.5,0.5,0.5,0.5,0.5,0.5,0.25,0.25,0.25,0.25,0.25,0.25,0,0,0,0,0,0,0.25,0,0.25,0,0.25])
mindist = 10000000
closestpoint = []
for i in range(len(x)):
    tdist = np.linalg.norm(tripod-x[i])
    if tdist < mindist:
        mindist = tdist
        closestpoint = i

print("ref point minimum distance: " + str(mindist))
print("point: " + np.array2string(x[closestpoint], formatter={'float_kind':lambda x: "%.2f" % x}))
print("score: " + str(data[closestpoint][7]))
```
